```python
import numpy as np
import cv2
import os
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
import random
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input  # ResNet50 imports

# Set random seed for reproducibility
random.seed(42)
np.random.seed(42)

folder_path = "trafficSignData"
label_file = 'labels.csv'
imageDimesions = (32, 32)  # For resizing
image_dimensions = (32, 32, 3)  # For ResNet50 input
ratio = 0.2

images = []
classNumber = []

# Load data
myList = os.listdir(folder_path)
print("View each classes in traffic folder :", myList, "\n")
lengthOfClasses = len(myList)
print("Total length of classes :", lengthOfClasses, "\n")

print("Iterating to load images & classes\n")
for x in range(0, lengthOfClasses):
    imagesList = os.listdir(os.path.join(folder_path, str(x)))
    for img in imagesList:
        currentImage = cv2.imread(folder_path + "/" + str(x) + "/" + img)
        if currentImage is not None:
            currentImage = cv2.resize(currentImage, imageDimesions)
            images.append(currentImage)
            classNumber.append(x)
        else:
            print(f"It is wrong folder path {folder_path}")
    print(x, end=" ")

# Convert to numpy arrays
images = np.array(images)
classNumber = np.array(classNumber)

print(images.shape)
print(classNumber.shape)

# Split data with fixed random state
X_train, X_test, y_train, y_test = train_test_split(images, classNumber, test_size=ratio, random_state=42)
print("After splitting for images for training:", X_train.shape)
print("After splitting for images for testing:", X_test.shape)
print("After splitting for classNumber for training & testing:", y_train.shape, y_test.shape)

X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, test_size=ratio, random_state=42)
print('Get the X_validation:', X_validation.shape)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
print(X_validation.shape, y_validation.shape)

# Assertions to verify data consistency
try:
    assert X_train.shape[0] == y_train.shape[0], "Number of images and class should match for training"
    assert X_test.shape[0] == y_test.shape[0], "Number of images and class should match for testing"
    assert X_train.shape[1:] == image_dimensions, "Dimensions should match for training"
    assert X_test.shape[1:] == image_dimensions, "Dimensions should match for testing"
    print("All tests passed")
except AssertionError as e:
    print(e)

# [Your existing visualization code for class distribution remains unchanged]

# Preprocessing for ResNet50 (RGB, no grayscale)
def preprocessing(img):
    img = preprocess_input(img)  # ResNet50-specific preprocessing
    return img

# Apply preprocessing
X_train = np.array(list(map(preprocessing, X_train)))
X_test = np.array(list(map(preprocessing, X_test)))
X_validation = np.array(list(map(preprocessing, X_validation)))

print("After preprocessing", X_train.shape)  # Should be (num_samples, 32, 32, 3)

# Data augmentation
dataGenerator = ImageDataGenerator(
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    shear_range=0.1,
    rotation_range=10
)
dataGenerator.fit(X_train)

# Visualize augmented images (updated for RGB)
batches = dataGenerator.flow(X_train, y_train, batch_size=20)
X_batch, y_batch = next(batches)
fig, axs = plt.subplots(1, 10, figsize=(20, 5))
fig.tight_layout()
for i in range(10):
    axs[i].imshow(X_batch[i])  # Remove cmap='gray' for RGB
    axs[i].axis('off')
plt.show()

# One-hot encode labels
y_train = to_categorical(y_train, lengthOfClasses)
y_validation = to_categorical(y_validation, lengthOfClasses)
y_test = to_categorical(y_test, lengthOfClasses)

# Define ResNet50 model
def myResNetModel():
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(image_dimensions[0], image_dimensions[1], 3))
    for layer in base_model.layers:
        layer.trainable = False  # Freeze pre-trained layers
    
    model = Sequential([
        base_model,
        GlobalAveragePooling2D(),
        Dense(512, activation='relu'),
        Dropout(0.5),
        Dense(256, activation='relu'),
        Dropout(0.5),
        Dense(lengthOfClasses, activation='softmax')  # Use lengthOfClasses directly
    ])
    
    model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Training parameters
batch_size = 50
steps_per_epoch = 1000
epochs = 10

# Train the model
model = myResNetModel()
print(model.summary())
history = model.fit(
    dataGenerator.flow(X_train, y_train, batch_size=batch_size),
    epochs=epochs,
    validation_data=(X_validation, y_validation),
    steps_per_epoch=steps_per_epoch,
    verbose=1
)

# Save the model
model.save('Traffic_Sign_Model_ResNet50.h5')

# Evaluate the model
model = load_model('Traffic_Sign_Model_ResNet50.h5')
score = model.evaluate(X_test, y_test, verbose=0)
print("Result of Loss Value:", score[0])
print("Result of Accuracy:", score[1])

# [Your existing plotting code for history remains unchanged]
fig, (ax1, ax2) = plt.subplots(2, figsize=(8, 6))
val_acc = history.history.get('val_accuracy', [])
val_loss = history.history.get('val_loss', [])
ax1.plot(val_acc, label='Validation Accuracy')
ax2.plot(val_loss, label='Validation Loss')
ax1.set_ylabel('Validation Accuracy', fontsize=14)
ax2.set_ylabel('Validation Loss', fontsize=14)
ax1.set_xlabel('Epochs', fontsize=14)
ax1.set_title('Validation Accuracy per Epoch', fontsize=16)
ax2.set_title('Validation Loss per Epoch', fontsize=16)
ax1.grid(True, linestyle='--', alpha=0.7)
ax2.grid(True, linestyle='--', alpha=0.7)
ax1.legend(fontsize=12, loc='upper left', title='Legend', title_fontsize=13)
ax2.legend(fontsize=12, loc='upper left', title='Legend', title_fontsize=13)
plt.tight_layout()
plt.show()
```
